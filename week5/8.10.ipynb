```python
from sklearn.datasets import fetch_20newsgroups
from gensim import corpora, models

data = fetch_20newsgroups(subset='train')
data = data['data']
data = data[:50]
```

```python
from pprint import pprint
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english'))

texts = [
     [word for word in document.lower().split() if word not in stopwords]
     for document in data
 ]
```

```python
#  создаем словарь 
dictionary = corpora.Dictionary(texts)
```

```python
# проводим векторизацию
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
# tfidf model
tfidf_model = models.TfidfModel(corpus)  
```

```python
# tfidf 
tfidf = tfidf_model[corpus]
```

Here we can implement
- RpModel
- LsiModel
- HdpModel
- LdaModel

```python
# init model
lsi = models.LdaModel(tfidf, id2word=dictionary, num_topics=20)
```

```python
corpus_lsi = lsi[tfidf]
```

```python
lsi.print_topics()
```

```python
lsi.id2word
```

```python
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lsi, corpus, dictionary=lsi.id2word)
vis
```
